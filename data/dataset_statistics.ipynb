```python
import csv
from collections import defaultdict

rows = []
sum_dict = defaultdict(int)
dose_dict = defaultdict(int)

with open("dataset1_200.csv") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)  # skip header
    for row in reader:
        row = list(map(float, row))
        row_sum = sum(row[2::2])
        sum_dict[row_sum] += 1
        dose_count = sum([1 for x in row[1:-1:2] if x > 0])
        dose_dict[dose_count] += 1
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.bar(*zip(*dose_dict.items()))
plt.ylabel('Number of protocols')
plt.xlabel('Doses count')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.bar(*zip(*sum_dict.items()))
plt.ylabel('Number of protocols')
plt.xlabel('Sum of doses')
plt.show()
```
